## Data cleaning

### Subtask:
Clean the three datasets: `df_historical`, `df_fear_greed`, and `df_equality`.


**Reasoning**:
Clean the three datasets according to the instructions, handling missing values, converting data types, and addressing outliers.



```python
# Clean df_historical
df_historical_cleaned = df_historical.copy()

# Handle missing values in df_historical
for col in df_historical_cleaned.columns:
    if df_historical_cleaned[col].dtype in ['int64', 'float64']:
        df_historical_cleaned[col] = df_historical_cleaned[col].fillna(df_historical_cleaned[col].median())
    elif df_historical_cleaned[col].dtype == 'object':
        df_historical_cleaned[col] = df_historical_cleaned[col].fillna(df_historical_cleaned[col].mode()[0])

# Convert 'Timestamp IST' to datetime
try:
    df_historical_cleaned['Timestamp IST'] = pd.to_datetime(df_historical_cleaned['Timestamp IST'], format='%d-%m-%Y %H:%M', errors='coerce')
    # Remove rows where conversion failed
    df_historical_cleaned = df_historical_cleaned.dropna(subset=['Timestamp IST'])
except Exception as e:
    print(f"Error converting 'Timestamp IST': {e}")

# Clean df_fear_greed
df_fear_greed_cleaned = df_fear_greed.copy()

# Double-check for missing values and handle them
for col in df_fear_greed_cleaned.select_dtypes(include=['number', 'object']):
    if df_fear_greed_cleaned[col].isnull().any():
        if df_fear_greed_cleaned[col].dtype in ['int64', 'float64']:
            df_fear_greed_cleaned[col] = df_fear_greed_cleaned[col].fillna(df_fear_greed_cleaned[col].median())
        elif df_fear_greed_cleaned[col].dtype == 'object':
            df_fear_greed_cleaned[col] = df_fear_greed_cleaned[col].fillna(df_fear_greed_cleaned[col].mode()[0])

# Convert 'date' and 'timestamp' to datetime
df_fear_greed_cleaned['date'] = pd.to_datetime(df_fear_greed_cleaned['date'])
df_fear_greed_cleaned['timestamp'] = pd.to_datetime(df_fear_greed_cleaned['timestamp'], unit='s')

# Clean df_equality
df_equality_cleaned = df_equality.copy()

# Drop 'Equality class' column due to 100% missing values
df_equality_cleaned = df_equality_cleaned.drop('Equality class', axis=1)

# Check for other missing values in df_equality
for col in df_equality_cleaned.columns:
    if df_equality_cleaned[col].isnull().any():
        if df_equality_cleaned[col].dtype in ['int64', 'float64']:
            df_equality_cleaned[col] = df_equality_cleaned[col].fillna(df_equality_cleaned[col].median())
        elif df_equality_cleaned[col].dtype == 'object':
            df_equality_cleaned[col] = df_equality_cleaned[col].fillna(df_equality_cleaned[col].mode()[0])
```

## Data analysis

### Subtask:
Analyze the relationships between variables in the cleaned datasets.


**Reasoning**:
Perform correlation analysis on the numerical features of df_historical_cleaned, time series analysis on df_historical_cleaned and df_fear_greed_cleaned, and analyze the df_equality_cleaned dataset, then visualize the results.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Correlation Analysis for df_historical_cleaned
numerical_cols = df_historical_cleaned.select_dtypes(include=['number']).columns
correlation_matrix = df_historical_cleaned[numerical_cols].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features (df_historical_cleaned)')
plt.show()

# Time Series Analysis
plt.figure(figsize=(15, 6))
plt.plot(df_historical_cleaned['Timestamp IST'], df_historical_cleaned['Closed PnL'], label='Closed PnL')
plt.xlabel('Timestamp IST')
plt.ylabel('Closed PnL')
plt.title('Time Series Plot of Closed PnL')
plt.legend()
plt.show()


plt.figure(figsize=(15, 6))
plt.plot(df_fear_greed_cleaned['timestamp'], df_fear_greed_cleaned['value'], label='Fear & Greed Index')
plt.xlabel('Timestamp')
plt.ylabel('Fear & Greed Index Value')
plt.title('Time Series Plot of Fear & Greed Index')
plt.legend()
plt.show()

#Equality Data Analysis
plt.figure(figsize=(10,6))
sns.boxplot(x='Factory', y='Equality Score', data=df_equality_cleaned)
plt.title('Equality Score Distribution by Factory')
plt.xticks(rotation=45, ha='right')
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(x='Job Role', y='Equality Score', data=df_equality_cleaned)
plt.title('Equality Score Distribution by Job Role')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## Data visualization

### Subtask:
Create visualizations to illustrate key findings from the data analysis.


**Reasoning**:
Generate the requested visualizations based on the cleaned dataframes.



```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)

# 1. Time Series Comparison
plt.figure(figsize=(15, 6))
plt.plot(df_historical_cleaned['Timestamp IST'], df_historical_cleaned['Closed PnL'], label='Closed PnL', color='blue')

# Find the closest dates in df_fear_greed_cleaned for each date in df_historical_cleaned
# Resample df_fear_greed_cleaned to daily frequency, selecting only numerical columns
df_fear_greed_resampled = df_fear_greed_cleaned[['timestamp', 'value']].resample('D', on='timestamp').mean()
# Merge dataframes
merged_df = pd.merge_asof(df_historical_cleaned.sort_values('Timestamp IST'),
                         df_fear_greed_resampled.reset_index(),
                         left_on='Timestamp IST',
                         right_on='timestamp',
                         direction='nearest')
plt.plot(merged_df['Timestamp IST'], merged_df['value'], label='Fear & Greed Index', color='red')

plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.title('Time Series Comparison of Closed PnL and Fear & Greed Index')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Correlation Heatmap Enhancement
numerical_cols = df_historical_cleaned[['Size Tokens', 'Size USD', 'Closed PnL', 'Fee', 'Execution Price']].select_dtypes(include=['number']).columns
correlation_matrix = df_historical_cleaned[numerical_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap (Enhanced)')
plt.show()

# 3. Equality Score Distribution (Enhanced)
plt.figure(figsize=(12, 6))
sns.violinplot(x='Factory', y='Equality Score', data=df_equality_cleaned, palette="Set3")
plt.title('Equality Score Distribution by Factory (Violin Plot)')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.violinplot(x='Job Role', y='Equality Score', data=df_equality_cleaned, palette="Set3")
plt.title('Equality Score Distribution by Job Role (Violin Plot)')
plt.xticks(rotation=45)
plt.show()

# 4. Combined Visualization
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Time Series Comparison
axes[0, 0].plot(merged_df['Timestamp IST'], merged_df['Closed PnL'], label='Closed PnL', color='blue')
axes[0, 0].plot(merged_df['Timestamp IST'], merged_df['value'], label='Fear & Greed Index', color='red')
axes[0, 0].set_xlabel('Timestamp')
axes[0, 0].set_ylabel('Value')
axes[0, 0].set_title('Time Series Comparison')
axes[0, 0].legend()
axes[0, 0].tick_params(axis='x', rotation=45)

# Correlation Heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", ax=axes[0, 1])
axes[0, 1].set_title('Correlation Heatmap')

# Equality Score Distribution by Factory
sns.violinplot(x='Factory', y='Equality Score', data=df_equality_cleaned, palette="Set3", ax=axes[1, 0])
axes[1, 0].set_title('Equality Score by Factory')
axes[1, 0].tick_params(axis='x', rotation=45)

# Equality Score Distribution by Job Role
sns.violinplot(x='Job Role', y='Equality Score', data=df_equality_cleaned, palette="Set3", ax=axes[1, 1])
axes[1, 1].set_title('Equality Score by Job Role')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
